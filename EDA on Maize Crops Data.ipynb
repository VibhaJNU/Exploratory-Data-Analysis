```python
#Model Building - Linear Regression and Decision Tree
```

```python
#Splitting the dataset
X= data.iloc[:,2:-1]
Y= data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr_reg= LinearRegression(normalize=True)

lr_reg.fit(X_train, Y_train)
```

```python
coff= pd.DataFrame(lr_reg.coef_, index=X_train.columns).sort_values(by=[0], ascending=False)
coff.rename(columns={0:'coff'})
```

```python
print('linear intercept is {}'.format(lr_reg.intercept_))
```

```python
#Evaluating metrics for linear regression
from sklearn import metrics
print('R2 score for linear reg is {}'.format(metrics.r2_score(Y_test, lr_reg.predict(X_test))))
print('MSE for linear reg is {}'.format(metrics.mean_squared_error(Y_test, lr_reg.predict(X_test))))
print('RMSE for linear reg is {}'.format(np.sqrt(metrics.mean_squared_error(Y_test,lr_reg.predict(X_test)))))
```

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor
dtree= DecisionTreeRegressor(criterion='mse', random_state=1)

dtree.fit(X_train, Y_train)
```

```python
print('R2 score for decision tree is {}'.format(metrics.r2_score(Y_test, dtree.predict(X_test))))
print('RMSE for decision tree is {}'.format(np.sqrt(metrics.mean_squared_error(Y_test, dtree.predict(X_test)))))
```

```python
pd.Series(dtree.feature_importances_, index=X_train.columns).sort_values(ascending=False)

sns.barplot(x=X_train.columns, y=dtree.feature_importances_)
plt.xticks(rotation=90)
plt.plot()
```

```python
#Looks like the Price is more dependent on = MSP, International Price Production and Area
#Lets build decision tree again taking only these 4 variables
X_tr = X_train[['MSP', 'International', 'Growth Rate', 'Area (million hactares)']]
X_te = X_test[['MSP', 'International', 'Growth Rate', 'Area (million hactares)']]
```

```python
dtree1= DecisionTreeRegressor(criterion='mse', random_state=2)
dtree1.fit(X_tr, Y_train)
```

```python
print('R2 score for decision tree is {}'.format(metrics.r2_score(Y_test, dtree1.predict(X_te))))
print('RMSE for decision tree is {}'.format(np.sqrt(metrics.mean_squared_error(Y_test, dtree1.predict(X_te)))))
```

**Conclusion**
Variables on which Domestic Price depend are = MSP, International Price, Production and Area
